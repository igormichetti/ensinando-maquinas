```python
X_linreg = data.GrLivArea
y = data.SalePrice
```

```python
from sklearn.model_selection import train_test_split
X_train_linreg, X_test_linreg, y_train, y_test = train_test_split(
    X_linreg, y, test_size=0.2)
```

```python
plt.plot(X_linreg, y, "b.")
plt.xlabel("Above Grade Living Area")
plt.ylabel("House Sale Price")
```

```python
X_b = np.c_[np.ones((len(X_train_linreg),1)), X_train_linreg]
lin_params = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y_train) #normal equation
lin_params
```

```python
X_predlin = np.array([[0], [3000]])
X_predlin_b = np.c_[np.ones((2,1)), X_predlin] # adds null vector for theta0
y_predlin = X_predlin_b.dot(lin_params)
y_predlin
```

```python
plt.plot(X_linreg, y, "b.")
plt.plot(X_predlin, y_predlin, "r-", label="Prediction Line")
plt.axis([0, 3000, 0, 500000])
plt.xlabel("Above Grade Living Area")
plt.ylabel("House Sale Price")
plt.legend(loc="upper left", fontsize=14)
```

## Given the parameters, you can predict the outputs:
- by vectorization function 
- by the dot product
- by sklearn LinearRegression class

```python
# Vectorization (iterate throght every feature and apply the line function)
def linregf(x):
    return lin_params[0] + ( x * lin_params[-1])
linreg_vectorizer = np.vectorize(linregf)
linreg_vect_pred = linreg_vectorizer(X_train_linreg)
print(linreg_vect_pred)
```
